```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
# True underlying function
# ------------------------
def signal(x):
  return np.cos(5 * x) + np.sin(3 * x)
```

```python
# Training data
# -------------
N = 50
x_train = np.linspace(0, 2 * np.pi, N)
y_train = signal(x_train) + np.random.normal(0, 0.4, N)

# True underlying function
# ------------------------
x_true = np.linspace(0, 2 * np.pi, 500)
y_true = signal(x_true)
```

```python
# Plot model vs. truth
# --------------------
plt.figure(figsize = (10, 5))
plt.scatter(x_train, y_train, label = 'Training data', c = 'black')
plt.plot(x_train, y_hat, c = 'orange', label = 'Learned model')
plt.plot(x_true, y_true, label = 'True function')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend(loc = 'lower left')
plt.title('RKHS Regression')
plt.savefig('RKHS_regression.png')
plt.show()
```
